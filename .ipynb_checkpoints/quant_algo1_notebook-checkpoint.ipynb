## Simple Moving Average

A simple moving average (SMA) calculates the average of a selected range of prices, usually closing prices, by the number of periods in that range.

$$
SMA= (A1+A2+A3...+An)/n \\
A=Asset Price At Closing \\
n=Number Of Periods \\
$$

We'll use Nasdaq to simulate how effective our algorithm might be. Using stock prices for the last 5 years, we create the 50-day and 200-day SMAs.

```python
ticker = "NDAQ"
stock = yf.Ticker(ticker)

hist_data = stock.history(period="5y", interval="1d")
hist_data.drop(["Volume", "Dividends", "Stock Splits"],
               inplace=True, axis=1)
hist_data.tail(10)
```

```python
hist_data["short_sma"] = hist_data['Close'].rolling(window = 50).mean()
hist_data["long_sma"] = hist_data['Close'].rolling(window = 200).mean()

hist_data[["short_sma","long_sma"]].tail(10)
```

```python
fig, ax = plt.subplots(figsize=(12,6))

def plot_hist_data(ax):
    ax.plot(hist_data.index, hist_data.loc[:, "Close"], label="Stock Price")
    ax.plot(hist_data.index, hist_data["short_sma"], label = "50 Day SMA")
    ax.plot(hist_data.index, hist_data["long_sma"], label = "200 Day SMA")

    ax.set_title(ticker)
    ax.legend(loc='best')
    ax.set_ylabel("Price ($)")

plot_hist_data(ax)
```

<a id='sec2'></a>
## The SMA Algorithm

The algorithm will use a 50-day SMA and a 200-day SMA for stock prices to initiate a strike price that will indicate a buy or sell. The strategy is that when the averages equal/intersect it will give an automatic buy/sell based on where the averages are trending.

Using the 5 year data, we will buy whenever the 50-day average becomes greater than the 200-day average and sell when the reverse occurs. Since this is the only stock in our portfolio, a buy will put 100%, or 1.0 of our assets in the stock, and a sell will have 0% investment. The trading history graph below reflects this.

```python
sma_diffs = hist_data["short_sma"] - hist_data["long_sma"]
sma_diffs = sma_diffs.replace(np.nan, 0)
sma_diffs.loc[sma_diffs > 0] = 1
sma_diffs.loc[sma_diffs <= 0] = 0

trading_pos = sma_diffs.shift(1)

fig1, (ax1, ax2) = plt.subplots(2, 1, figsize = (12, 6))
plot_hist_data(ax1)
ax2.plot(trading_pos, label="Trading Position")
ax2.set_title("Trading History")
```

```python
trading_hist = trading_pos[trading_pos.diff() != 0][2:].rename("Action").to_frame()
trading_hist = trading_hist.replace(1, "Buy")
trading_hist = trading_hist.replace(0, "Sell")
trading_hist["Price"] = hist_data["Close"]
trading_hist = trading_hist.append(pd.DataFrame([[hist_data.index[-1], "Sell", hist_data["Close"][-1]]], columns = ["Date", "Action", "Price"]).set_index("Date"))
trading_hist
```

<a id='sec3'></a>
## Performance

Now that we know what our investment strategy looks like, we will see how it actually performs against a simple Buy/Hold. Again, we will use the data we have collected from $NDAQ.

```python
def get_roi(df):
    roi = 1
    for action, price in df.itertuples(index = False):
        if action == "Buy": cost = price
        if action == "Sell":
            roi = (price/cost) * roi
    return roi * 100

# hist_data["Close"]
# trading_pos[]
# trading_hist.index[0]

sma_roi = get_roi(trading_hist)
buy_hold_roi = trading_hist["Price"][-1] / trading_hist["Price"][0] * 100

print("Buy/Hold ROI: " + str(round(buy_hold_roi, 2)) + "%")
print("SMA ROI: " + str(round(sma_roi, 2)) + "%")
```
